Aplicación de algoritmo Binary Insertion Sort para la Columna ISSN

```python
# Función para preparar los datos de la columna 'ISSN'
def preparar_datos_ISSN(df, column_name='ISSN'):
    # Calcular el total de caracteres y elementos antes de la preparación
    total_caracteres = df[column_name].astype(str).str.len().sum()
    total_elementos = df[column_name].count()

    print(f"Total de caracteres en la columna '{column_name}': {total_caracteres}")
    print(f"Total de elementos en la columna '{column_name}': {total_elementos}")

    # Convertir los datos de la columna a numéricos, si es posible
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce').fillna(0).astype(int)
    
    return df

# Implementación del algoritmo Binary Insertion Sort
def binary_search(arr, val, start, end):
    # Buscar la posición correcta para insertar 'val' en 'arr'
    if start == end:
        if arr[start] > val:
            return start
        else:
            return start + 1
    if start > end:
        return start

    mid = (start + end) // 2
    if arr[mid] < val:
        return binary_search(arr, val, mid + 1, end)
    elif arr[mid] > val:
        return binary_search(arr, val, start, mid - 1)
    else:
        return mid

def binary_insertion_sort(arr):
    for i in range(1, len(arr)):
        val = arr[i]
        j = binary_search(arr, val, 0, i - 1)
        arr = arr[:j] + [val] + arr[j:i] + arr[i+1:]
    return arr

# Adaptar Binary Insertion Sort para ordenar la columna 'ISSN'
def ordenar_binary_insertion_ISSN(df, column_name='ISSN'):
    # Convertir la columna a lista
    issn_list = df[column_name].tolist()

    # Ordenar usando Binary Insertion Sort
    issn_list = binary_insertion_sort(issn_list)

    # Asignar el resultado ordenado a la columna del DataFrame
    df[column_name] = issn_list
    return df

# Preparar los datos de la columna 'ISSN'
df_preparado = preparar_datos_ISSN(df.copy())

# Ejecutar el proceso 4 veces y calcular los tiempos
execution_times = []
for _ in range(4):
    start_time = time.time()
    df_sorted = ordenar_binary_insertion_ISSN(df_preparado.copy())  # Hacer una copia del DataFrame preparado
    end_time = time.time()
    
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Calcular el tiempo promedio
average_time = sum(execution_times) / len(execution_times)

# Imprimir resultados
print("Tiempos de ejecución:", execution_times)
print("Tiempo promedio:", average_time)
```

Aplicación de algoritmo Binary Insertion Sort para la Columna fecha publicacion

```python
# Función para preparar los datos de la columna 'Fecha Publicacion' tomando solo el año
def preparar_datos_fecha(df, column_name='Fecha Publicacion'):
    # Extraer el año de cada fecha, manejando valores nulos
    df[column_name] = pd.to_datetime(df[column_name], errors='coerce').fillna(pd.Timestamp.min).dt.year

    # Calcular el total de caracteres y elementos en la columna
    total_caracteres = df[column_name].astype(str).str.len().sum()
    total_elementos = df[column_name].count()

    print(f"Total de caracteres en la columna '{column_name}': {total_caracteres}")
    print(f"Total de elementos en la columna '{column_name}': {total_elementos}")

    return df

# Función de búsqueda binaria para encontrar la posición de inserción
def binary_search(arr, val, start, end):
    if start == end:
        if arr[start] > val:
            return start
        else:
            return start + 1
    if start > end:
        return start

    mid = (start + end) // 2
    if arr[mid] < val:
        return binary_search(arr, val, mid + 1, end)
    elif arr[mid] > val:
        return binary_search(arr, val, start, mid - 1)
    else:
        return mid

# Implementación del algoritmo Binary Insertion Sort
def binary_insertion_sort(arr):
    for i in range(1, len(arr)):
        val = arr[i]
        # Solo proceder si el valor no es nulo
        if pd.notna(val):
            j = binary_search(arr, val, 0, i - 1)
            arr = arr[:j] + [val] + arr[j:i] + arr[i + 1:]
    return arr

# Adaptar Binary Insertion Sort para ordenar la columna 'Fecha Publicacion'
def ordenar_binary_insertion_fecha(df, column_name='Fecha Publicacion'):
    # Convertir la columna a lista, eliminando valores nulos
    fechas_list = df[column_name].dropna().tolist()
    
    # Ordenar usando Binary Insertion Sort
    fechas_list = binary_insertion_sort(fechas_list)
    
    # Asignar el resultado ordenado a la columna del DataFrame
    # Reemplazar los valores originales en el DataFrame
    df[column_name] = pd.Series(fechas_list + [None] * (len(df) - len(fechas_list)), index=df.index)
    return df

# Preparar los datos de la columna 'Fecha Publicacion'
df_preparado = preparar_datos_fecha(df.copy())

# Ejecutar el proceso 4 veces y calcular los tiempos
execution_times = []
for _ in range(4):
    start_time = time.time()
    df_sorted = ordenar_binary_insertion_fecha(df_preparado.copy())  # Hacer una copia del DataFrame preparado
    end_time = time.time()
    
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Calcular el tiempo promedio
average_time = sum(execution_times) / len(execution_times)

# Imprimir resultados
print("Tiempos de ejecución:", execution_times)
print("Tiempo promedio:", average_time)
```
