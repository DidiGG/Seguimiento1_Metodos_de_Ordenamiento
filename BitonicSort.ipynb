Aplicación de algoritmo Bitonic Sort para la Columna ISSN

```python
# Función para preparar los datos de la columna 'ISSN'
def preparar_datos_ISSN(df, column_name='ISSN'):
    # Calcular el total de caracteres y elementos antes de la preparación
    total_caracteres = df[column_name].astype(str).str.len().sum()
    total_elementos = df[column_name].count()

    print(f"Total de caracteres en la columna '{column_name}': {total_caracteres}")
    print(f"Total de elementos en la columna '{column_name}': {total_elementos}")

    # Convertir los datos de la columna a numéricos, si es posible
    df[column_name] = pd.to_numeric(df[column_name], errors='coerce').fillna(0).astype(int)
    
    return df

# Función para comparar e intercambiar elementos
def compAndSwap(a, i, j, dire):
    if (dire == 1 and a[i] > a[j]) or (dire == 0 and a[i] < a[j]):
        a[i], a[j] = a[j], a[i]

# Función para la fusión bitónica
def bitonicMerge(a, low, cnt, dire):
    if cnt > 1:
        k = cnt // 2
        for i in range(low, low + k):
            compAndSwap(a, i, i + k, dire)
        bitonicMerge(a, low, k, dire)
        bitonicMerge(a, low + k, k, dire)

# Función recursiva para ordenar secuencias bitónicas
def bitonicSort(a, low, cnt, dire):
    if cnt > 1:
        k = cnt // 2
        bitonicSort(a, low, k, 1)    # Ordenar la primera mitad en orden ascendente
        bitonicSort(a, low + k, k, 0)  # Ordenar la segunda mitad en orden descendente
        bitonicMerge(a, low, cnt, dire)

# Función para iniciar la ordenación bitónica
def sort(a, N, up):
    bitonicSort(a, 0, N, up)

# Adaptar Bitonic Sort para ordenar la columna 'ISSN'
def ordenar_bitonic_ISSN(df, column_name='ISSN'):
    # Convertir la columna a lista
    issn_list = df[column_name].tolist()
    
    # Ordenar usando Bitonic Sort
    sort(issn_list, len(issn_list), 1)  # 1 significa orden ascendente
    
    # Asignar el resultado ordenado a la columna del DataFrame
    df[column_name] = issn_list
    return df

# Preparar los datos de la columna 'ISSN'
df_preparado = preparar_datos_ISSN(df.copy())

# Ejecutar el proceso 4 veces y calcular los tiempos
execution_times = []
for _ in range(4):
    start_time = time.time()
    df_sorted = ordenar_bitonic_ISSN(df_preparado.copy())  # Hacer una copia del DataFrame preparado
    end_time = time.time()
    
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Calcular el tiempo promedio
average_time = sum(execution_times) / len(execution_times)

# Imprimir resultados
print("Tiempos de ejecución:", execution_times)
print("Tiempo promedio:", average_time)
```

Aplicación de algoritmo Bitonic Sort para la Columna para fecha de pubicación

```python

# Función para extraer el año de la columna 'Fecha Publicacion'
def extraer_año(fecha):
    # Utilizar una expresión regular para extraer el año (4 dígitos)
    match = re.search(r'(\d{4})', str(fecha))
    if match:
        return int(match.group(1))  # Devolver solo el año
    else:
        return pd.Timestamp.min.year  # Si no hay año válido, devolver el año mínimo (1)

# Función para preparar los datos de la columna 'Fecha Publicacion'
def preparar_datos_fecha(df, column_name='Fecha Publicacion'):
    # Calcular el total de caracteres y elementos antes de la preparación
    total_caracteres = df[column_name].astype(str).str.len().sum()
    total_elementos = df[column_name].count()

    print(f"Total de caracteres en la columna '{column_name}': {total_caracteres}")
    print(f"Total de elementos en la columna '{column_name}': {total_elementos}")

    # Aplicar la función para extraer el año
    df[column_name] = df[column_name].apply(extraer_año)
    
    return df

# Función para comparar e intercambiar elementos
def compAndSwap(a, i, j, dire):
    if (dire == 1 and a[i] > a[j]) or (dire == 0 and a[i] < a[j]):
        a[i], a[j] = a[j], a[i]

# Función para la fusión bitónica
def bitonicMerge(a, low, cnt, dire):
    if cnt > 1:
        k = cnt // 2
        for i in range(low, low + k):
            compAndSwap(a, i, i + k, dire)
        bitonicMerge(a, low, k, dire)
        bitonicMerge(a, low + k, k, dire)

# Función recursiva para ordenar secuencias bitónicas
def bitonicSort(a, low, cnt, dire):
    if cnt > 1:
        k = cnt // 2
        bitonicSort(a, low, k, 1)    # Ordenar la primera mitad en orden ascendente
        bitonicSort(a, low + k, k, 0)  # Ordenar la segunda mitad en orden descendente
        bitonicMerge(a, low, cnt, dire)

# Función para iniciar la ordenación bitónica
def sort(a, N, up):
    bitonicSort(a, 0, N, up)

# Adaptar Bitonic Sort para ordenar la columna 'Fecha Publicacion'
def ordenar_bitonic_fecha(df, column_name='Fecha Publicacion'):
    # Convertir la columna a lista
    fecha_list = df[column_name].tolist()
    
    # Ordenar usando Bitonic Sort
    sort(fecha_list, len(fecha_list), 1)  # 1 significa orden ascendente
    
    # Asignar el resultado ordenado a la columna del DataFrame
    df[column_name] = fecha_list
    return df

# Preparar los datos de la columna 'Fecha Publicacion'
df_preparado = preparar_datos_fecha(df.copy())

# Ejecutar el proceso 4 veces y calcular los tiempos
execution_times = []
for _ in range(4):
    start_time = time.time()
    df_sorted = ordenar_bitonic_fecha(df_preparado.copy())  # Hacer una copia del DataFrame preparado
    end_time = time.time()
    
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Calcular el tiempo promedio
average_time = sum(execution_times) / len(execution_times)

# Imprimir resultados
print("Tiempos de ejecución:", execution_times)
print("Tiempo promedio:", average_time)
```
